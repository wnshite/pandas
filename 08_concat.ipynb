## concat()

- 단순 나열
- 똑같은 데이터 유형의 파일을 나열할 때

```python
# pd.concat([df1, df2]).iloc[170:190]
pd.concat([df1, df2], ignore_index=True)      #위아래로 연결
```

```python
sales1 = df1.iloc[:, :5]
sales2 = df2.iloc[:, 5:]
```

```python
sales1.info()
sales2.info()
```

```python
pd.concat([sales1, sales2], axis=1)   ##하나로 합치는 axis
```

## merge()

- 서로 다른 데이터 유형을 병합할 때
- 중심축으로 column이 다를때 사용

```python
info = {
    '고객명': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
df1
```

```python
df2
```

```python
pd.merge(df1, df2)   #교집합만 남게 만드는게 merge의 역할
```

```python
pd.merge(df1, df2, how='inner')    ## 교집합만
```

```python
pd.merge(df1, df2, how='outer')   ## 전체 데이터 출력, 없는 값은 NaN으로 출력
```

```python
pd.merge(df1, df2, how='left')
```

```python
pd.merge(df1, df2, how='right')
```

```python
info = {
    '이름': ['박세리', '이대호', '손흥민', '김연아', '마이클조던'],
    '생년월일': ['1980-01-02', '1982-02-22', '1993-06-12', '1988-10-16', '1970-03-03'],
    '성별': ['여자', '남자', '남자', '여자', '남자']
}

money = {
    '고객명': ['김연아', '박세리', '손흥민', '이대호', '타이거우즈'],
    '연봉': ['2000원', '3000원', '1500원', '2500원', '3500원']
}
```

```python
df1 = pd.DataFrame(info)
df2 = pd.DataFrame(money)
```

```python
pd.merge(df1, df2, left_on='이름', right_on='고객명')   ## 다른 두 데이터를 하나로 병합
```
