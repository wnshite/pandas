```python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')
```

```python
df.head(3)
df_copy = df.copy()
```

## groupby()

```python
df_copy.groupby('sex').mean(numeric_only=True)
```

```python
df_copy.groupby(['sex', 'pclass']).mean(numeric_only=True)   ### 데이터를 뭉쳐서 보는 개념
```

```python
df_copy.groupby(['sex', 'pclass'])['survived'].mean(numeric_only=True)   ## survived만 보고 싶을때 추가로 []적기
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)   ## 리스트로 한번 더 감싸면 예쁘게(?) 데이터 프레임으로 나옴.
```

## reset_index()

```python
df_temp = df_copy.groupby(['sex', 'pclass'])[['survived']].mean(numeric_only=True)
df_temp
```

```python
df_temp.reset_index()   # 묶어놨던 거 다시 쪼개기
```

```python
df_copy.groupby(['sex', 'pclass'])[['survived', 'age']].agg(['mean', 'sum'])   ## groupby에 groupby를 한 형태
```

## pivot_table

```python
df_copy.pivot_table(index='who', values='survived')
```

```python
df_copy.pivot_table(index=['who', 'pclass'], values='survived')
```

```python
df_copy.pivot_table(['sex'])[['survived']].mean()
```
